## 1. Creating your own function: Cartesian to Polar coordinates.


```python
from numpy import arctan
from math import sqrt,pi
x = int(input("x?"))
y = int(input("y?"))
def polar(x,y):
    r = sqrt((x**2)+(y**2))
    if x == 0:
        theta = 90
    else:
        theta = (arctan(y/x))*(180/pi)
    return print(f"r ={r:.3f} theta ={theta:.3f}")

polar(x,y)
```

## 2. Cartesian to Polar coordinates revisited 

```python
from numpy import arctan
from math import sqrt,pi
x = int(input("x?"))
y = int(input("y?"))
def polar(x,y):
    r = sqrt((x**2)+(y**2))
    if x == 0:
        theta = 90
    else:
        theta = (arctan(y/x))*(180/pi)
        if x<0 and y>0:
            theta += 180
        elif x<0 and y<0:
            theta = theta-180 
    return print(f"r ={r:.3f} theta ={theta:.3f}")

polar(x,y)
```

## 3. Newman 3.2: Parametric/polar plots of a curve

```python
import numpy as np
import matplotlib.pyplot as plt 
from numpy import cos,sin
from math import sqrt,pi
X = []
Y = []
for theta in np.linspace(0,2*pi):
    x = 2*cos(theta)+ cos(2*theta)
    y = 2*sin(theta) - sin(2*theta)
    X.append(x)
    Y.append(y)
plt.plot(X,Y)
plt.show()
```

## 4. Multiple plots in one figure.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos,sin
from math import sqrt,pi,exp

x1 = np.linspace(0,10,1000)
figure, axes = plt.subplots(nrows=2,ncols=2)


axes [0,0]. plot(x1,sin(x1))
axes [0,0]. set_title("sin(x)")
axes [0,1]. plot(x1,cos(x1))
axes [0,1]. set_title("cos(x)")
axes [1,0]. plot(x1,cos(x1)+x1)
axes [1,0]. set_title("cos(x)+x")
axes [1,1]. plot(x1,sin(x1)+x1)
axes [1,1]. set_title("sin(x)+x")

figure.tight_layout()
```

## 9. Using and re-using a function.

```python
import numpy as np
x= np.linspace(-10, 10, num=1000)
def equ1(x):
    return x+2*np.cos(x**2)
a = equ1(x)
plt.plot(a)
plt.show()
```

```python
import numpy as np
x= np.linspace(-10, 10, num=1000)
def equ1(x):
    return x+2*np.cos(x**2)
a = equ1(x)
b= equ1(x)+10
c= equ1(x+5)
plt.plot(a,label="a")
plt.plot(b,label="b")
plt.plot(c,label="c")
plt.legend()
plt.show()
```

```python
import numpy as np
x= np.linspace(-10, 10, num=1000)
def equ1(x):
    return abs(x)+2*np.cos(x**2)
a = equ1(x)
b= equ1(x)+10
c= equ1(x+5)
plt.plot(a,label="a")
plt.plot(b,label="b")
plt.plot(c,label="c")
plt.legend()
plt.show()
```
